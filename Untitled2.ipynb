```python
%matplotlib inline
import numpy, scipy, matplotlib.pyplot as plt, IPython.display as ipd
import librosa, librosa.display
import stanford_mir; stanford_mir.init()
```

```python
x, sr = librosa.load('audio/c_strum.wav')
ipd.Audio(x, rate=sr)
```

```python
plt.figure(figsize=(15, 5))
librosa.display.waveplot(x, sr, alpha=0.8)
```

```python
T = 2.0
f0 = 1047.0
sr = 22050
t = numpy.linspace(0, T, int(T*sr), endpoint=False)
x = 0.1*numpy.sin(2*numpy.pi*f0*t)
ipd.Audio(x, rate=sr)
```

```python
X = scipy.fft(x[:4096])
X_mag = numpy.absolute(X)
f = numpy.linspace(0, sr, 4096)
plt.figure(figsize=(14, 5))
plt.plot(f[:2000], X_mag[:2000])
plt.xlabel("Frequency (Hz)")
```
